```python
import pandas as pd
## 讀取檔案
dt = pd.read_csv('adult.data',sep=", ",header=None,engine='python')
tt = pd.read_csv('adult.test',sep=", ",header=None,engine='python', skiprows=1)
dt
tt
```

```python
## 命名欄位
dt.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','label']
tt.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','label']
feature_cols = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country']
dt
tt
```

```python
for column in dt.columns:
    print(dt[column].value_counts())
    
for column in tt.columns:
    print(tt[column].value_counts())    
    
## education跟education-num數量一樣所以取education-num即可，故刪除education
dt = dt.drop(columns=['education'])
tt = tt.drop(columns=['education'])
```

```python
dt['label'] = dt['label'].astype('int64')
tt['label'] = tt['label'].astype('int64')
```

```python
dt = dt[dt['workclass']!='?']        
dt = dt[dt['occupation']!='?']       
dt = dt[dt['native-country']!='?'] 

tt = tt[tt['workclass']!='?']       
tt = tt[tt['occupation']!='?']       
tt = tt[tt['native-country']!='?']   
```

```python
# 確認無缺失值
for col,cnt in zip(dt.columns,(dt=='?').sum(axis = 0)):
    if cnt == 0:
        print(str(col) + ': ' + str(cnt) + ' records')
        
for col,cnt in zip(tt.columns,(tt=='?').sum(axis = 0)):
    if cnt == 0:
        print(str(col) + ': ' + str(cnt) + ' records')
```

```python
# feature_cols = dt.columns.drop(['label','age','education'])
dt = pd.get_dummies(dt)
# df.columns
feature_cols = dt.columns

print(*feature_cols,sep='\n')
print(dt.info())

# feature_cols = df.columns.drop(['label','age','education'])
tt = pd.get_dummies(tt)
# df.columns
feature_cols = tt.columns

print(*feature_cols,sep='\n')
print(dt.info())
print(tt.info())
```

```python
for column in dt.columns:
    if column not in tt.columns:
        tt[column] = 0
        print(dt[column])
        print(tt[column])
        break
```

```python
dX = dt.drop('label',axis=1)
dy = dt['label']   
# X = feature_cols
# y_name = ["label_<=50K","label_>50K"]

tX = tt.drop('label',axis=1)
tX.columns = dX.columns
ty = tt['label']   
print(dX)
print(dy)
print(tX)
print(ty)
```

```python
def createDataSetAdult():
    dataset = []
    labels = []
    for i in range(88): labels.append(str('A'+str(i)))
    for line in dt:
        strlist = line[:-2].split(" ")
        print(strlist,sep=",")
        dataset.append(list(str(strlist)))
    return dataset, labels
createDataSetAdult()
```

```python
from sklearn.model_selection import train_test_split
dX_train, dX_test, dy_train, dy_test = train_test_split(dX, dy, test_size=0.2, random_state=42) # 80% training and 20% test
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Create arrays to store scores
dt_scores = []
tt_scores = []

# Loop through max_depth values
max_depth_values = range(1,20)
for max_depth in max_depth_values:
    dt = DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, max_leaf_nodes=92)
    dt.fit(dX_train, dy_train)
    dt_score = dt.score(dX_test, dy_test)
    dt_scores.append(dt_score)
    
    tt = DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, max_leaf_nodes=92)
    tt.fit(tX, ty)
    tt_score = tt.score(tX, ty)
    tt_scores.append(tt_score)

# Plot the results
print(dt_scores)
print(tt_scores)
plt.figure(figsize=(10, 6))
plt.plot(max_depth_values, dt_scores, label='train data')
plt.plot(max_depth_values, tt_scores, label='test data')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('Decision Tree Accuracy vs. max_depth')
plt.xticks(max_depth_values) 
plt.legend()
plt.show()
```

```python
print("=======training_data=======")
train_best = max(dt_scores)
for i in range(len(dt_scores)):
    if dt_scores[i]==train_best:
        print(f' depth : {i} , acc : {train_best:.3f}')
        break
        
print("=======test_data=======")
test_best = max(tt_scores)
for i in range(len(dt_scores)):
    if tt_scores[i]==test_best:
        print(f' depth : {i} , acc : {test_best:.3f}')
        break
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Create arrays to store scores

ax = []
dt_scores = []
tt_scores = []

# Loop through max_depth values
for node in range(2,100):
    ax.append(node)
    dt = DecisionTreeClassifier(criterion='entropy', max_depth=14, max_leaf_nodes = node)
    dt.fit(dX_train, dy_train)
    dt_scores.append(dt.score(dX_test, dy_test))
    tt_scores.append(dt.score(tX, ty))

# Plot the results
print(dt_scores)
print(tt_scores)
plt.figure(figsize=(10, 6))
plt.plot(ax, dt_scores, label='train data')
plt.plot(ax, tt_scores, label='test data')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('Decision Tree Accuracy vs. max_depth')
plt.xticks(ax) 
plt.legend()
plt.show()
```

```python
mas = max(tt_scores)
for i in range(len(tt_scores)):
    if tt_scores[i]==mas:
        print(i+1)
```
