```python
train_df = train_df[train_df['workclass']!='?']       
train_df = train_df[train_df['occupation']!='?']       
train_df = train_df[train_df['native-country']!='?']  
#刪除test有遺失值的data
test_df = test_df[test_df['workclass']!='?']       
test_df = test_df[test_df['occupation']!='?']      
test_df = test_df[test_df['native-country']!='?'] 

print(len(train_df))   #查看筆數
print(len(test_df))   #查看筆數
```

```python
#88 --> 87
train_df = pd.get_dummies(train_df)
test_df = pd.get_dummies(test_df)
feature_cols = train_df.columns
# print(*feature_cols,sep='\n')
train_df['label'] = train_df['label'].astype('int64')
test_df['label'] = test_df['label'].astype('int64')
for column in train_df.columns:
    if column not in test_df.columns:
        test_df[column] = 0
        print(train_df[column])
        print(test_df[column])
        break
print(train_df.info())
print(test_df.info())
```

```python
dX = train_df.drop('label',axis=1)
dy = train_df['label']   


tX = test_df.drop('label',axis=1)
tX.columns = dX.columns
ty = test_df['label'] 
print(dX)
print(dy)
```

```python
from sklearn.model_selection import train_test_split
dX_train, dX_test, dy_train, dy_test = train_test_split(dX, dy, test_size=0.2, random_state=42) # 80% training and 20% test
```

```python
dX_train
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import matplotlib.pyplot as plt

ax = []
a_train = []
a_test = []

for depth in range(1, 19):
    ax.append(depth)
    dt = DecisionTreeClassifier(criterion='gini', max_depth=depth, max_leaf_nodes=92)
    dt.fit(dX_train, dy_train)
    
    # 計算訓練集和測試集的準確度
    train_score = dt.score(dX_train, dy_train)
    test_score = dt.score(tX, ty)
    
    a_train.append(train_score)
    a_test.append(test_score)

print("=======training_data=======")
train_best_depth = np.argmax(a_train)
train_best_acc = max(a_train)
print(f"Best Depth for Maximum Training Accuracy: {train_best_depth}, Accuracy: {train_best_acc:.3f}")

print("=======test_data=======")
test_best_depth = np.argmax(a_test)
test_best_acc = max(a_test)
print(f"Best Depth for Maximum Test Accuracy: {test_best_depth}, Accuracy: {test_best_acc:.3f}")

# 繪製準確度曲線
plt.plot(ax, a_train, label='train')
plt.plot(ax, a_test, label='test')
plt.legend()
plt.xticks()
plt.show()
```
